## Why Matrix / Grid Plots?

Matrix or grid-style plots are perfect for:
- Showing correlations between variables  
- Visualizing pivot tables  
- Discovering patterns in large datasets


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()

tips.corr(numeric_only=True)
```

### Heatmaps (heatmap())

```python

sns.heatmap(tips.corr(numeric_only=True), annot=True)
plt.show()
```

### Cluster Map (clustermap())

```python
sns.clustermap(tips.corr(numeric_only=True), annot=True)
plt.show()
```
