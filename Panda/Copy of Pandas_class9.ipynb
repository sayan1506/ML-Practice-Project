```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Math': [85, 72, 91, 68, 79],
    'Science': [78, 65, 88, 70, 82]
}

df = pd.DataFrame(data)
print(df)
```

## 1. Applying Custom Functions with `apply()` / `map()`
- `apply()` lets you apply a custom function across rows or columns.
- `map()` is for applying functions element-wise to a Series.
- Both are handy for transformations, calculations, and cleaning data.

```python
df['Total Marks'] = df[['Math','Science']].apply(lambda x: x.sum(), axis=1)
print(df)
```

```python
col_mean = df[['Math','Science']].apply(lambda x: x.mean(), axis=0)
print(col_mean)
#
```

```python
def grade(s):
  if s >= 90:
    return "A"
  elif s >= 80:
    return "B"
  elif s >= 70:
    return "C"
  elif s >= 60:
    return "D"
  else:
    return "F"
df['Grade'] = df['Total Marks'].map(grade)
print(df)
```
