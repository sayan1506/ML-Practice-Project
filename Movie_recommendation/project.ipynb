```python
df['belongs_to_collection'].fillna('No Collection', inplace=True)
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df_new = df[['title', 'overview', 'genres', 'vote_average', 'popularity', 'tagline']]
df_new.head()
```

```python
df_new.isna().sum()
```

```python
df_new = df_new.dropna(subset=['title'])
```

```python
df_new['overview'].fillna('', inplace=True)
```

```python
df_new.iloc[0]['genres']
```

```python
import ast
df_new['genres'] = df_new['genres'].apply(lambda x : ' '.join([i["name"] for i in ast.literal_eval(x)]))
```

```python
df_new['genres']
```

```python
df_new['tagline'].fillna('', inplace=True)
```

```python
df_new['tags'] = df_new['overview'] + ' ' + df_new['genres'] + ' ' + df_new['tagline']
```

```python
df_new.head()
```

```python
df_new['tags'][1]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import  WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    # Remove special characters and digits
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenize the text
    words = text.split()
    # Remove stopwords and lemmatize
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(words)
```

```python
df_new['tags'] = df_new['tags'].apply(preprocess_text)
df_new['tags'].head()
```

```python
df_new = df_new.reset_index(drop=True)
```

```python
indices = pd.Series(df_new.index, index=df_new['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df_new['tags'])
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  # Handle duplicate titles by taking the first occurrence
  if isinstance(idx, pd.Series):
    idx = idx.iloc[0]
  
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df_new['title'].iloc[similar_idx]

```

```python
recommend('Kuch Kuch Hota Hai',5)
```
